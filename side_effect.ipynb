```python
import matplotlib.pyplot as plt #verileri gorsellestirmek icin gerekli kutuphane
```

```python
gender_distribution=df['Cinsiyet'].value_counts() # yan etkiye ugramis hastalarin cinsiyet dagilimi

plt.figure(figsize=(4,4)) #width and height respectively.
gender_distribution.plot(kind='bar',color=['pink','blue'])

plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

```

```python

city_distribution=df['Il'].value_counts() # Yan etkiye ugramis hastalarin yasadiklari illerin dagilimi

plt.figure(figsize=(4,4))
city_distribution.plot(kind="pie",autopct=lambda p: f'{int(p*sum(city_distribution)/100)}',cmap='Set3',legend=False)
plt.title('City Distribution')
plt.tight_layout()
plt.ylabel(" ")
plt.show()
```

```python
side_effect_distribution=df['Yan_Etki'].value_counts() # yan etki dagilimi

plt.figure(figsize=(15,6)) #width and height respectively.
side_effect_distribution.plot(kind='bar')

plt.title('Side Effect Distribution')
plt.xlabel('Side Effects')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
# yan etkiye ugramis kisilerin yaslara gore dagilimi

df['Age']=(pd.Timestamp.today()-df['Dogum_Tarihi']).dt.days/365.25

plt.figure(figsize=(4,4))
df['Age'].plot(kind="hist",bins=15,color='lightblue',edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()

plt.show()

```

```python
# Yan etki turlerine gore kisilerin yas dagilimlari
plt.figure(figsize=(15,10))
df.boxplot(column='Age',by='Yan_Etki',grid=False,vert=False)
plt.title('Age Distribution by Side Effect')
plt.suptitle(" ")
plt.xlabel('Age')
plt.ylabel('Side Effect')
plt.tight_layout()

plt.show()
```

```python
# Her ilaca ait bildirilen yan etki sayilari.
side_effect_per_drug=df.groupby('Ilac_Adi')['Yan_Etki'].count().sort_values(ascending=True)

plt.figure(figsize=(30,6))
side_effect_per_drug.plot(kind='bar',color='green')
plt.title('Number of Side Effects Per Drug')
plt.xlabel('Drug Name')
plt.ylabel('Number of Side Effects')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
#Her ilacin sahip oldugu farkli yan etki sayilari
different_side_effect_per_drug=df.groupby('Ilac_Adi')['Yan_Etki'].nunique().sort_values(ascending=True)

plt.figure(figsize=(30,6))
different_side_effect_per_drug.plot(kind='bar',color='green')
plt.title('Number of Different Side Effects Per Drug')
plt.xlabel('Drug Name')
plt.ylabel('Number of Different Side Effects')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python

#Yan Etki kolonu ile Cinsiyet Kolonunun birbiri ile iliskisinin arastirilmasi.
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['Yan_Etki'], df['Cinsiyet'])
chi2_stat, p_val, dof, ex = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2_stat:.2f}")
print(f"P-Value: {p_val:2f}")
```

```python
#Ilac Baslangic Tarihi, Ilac Bitis Tarihi ve Yan Etki Bildirim Tarihlerinin birbirleri ve Yan Etki Uzerinedeki Korelasyonun incelenmesi.
df['Start_Timestamp'] = (df['Ilac_Baslangic_Tarihi'] - df['Ilac_Baslangic_Tarihi'].min()).dt.days
df['End_Timestamp'] = (df['Ilac_Bitis_Tarihi'] - df['Ilac_Baslangic_Tarihi'].min()).dt.days
df['Effect_Timestamp'] = (df['Yan_Etki_Bildirim_Tarihi'] - df['Ilac_Baslangic_Tarihi'].min()).dt.days
```

```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import KNNImputer
```

```python
label_encoder = LabelEncoder()
df['Yan_Etki_Encoded'] = label_encoder.fit_transform(df['Yan_Etki'])


correlation_matrix = df[['Start_Timestamp', 'End_Timestamp', 'Effect_Timestamp', 'Yan_Etki_Encoded']].corr()
df.drop(columns=['Start_Timestamp','End_Timestamp','Effect_Timestamp','Yan_Etki_Encoded'],inplace=True)
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Date Columns and  Yan Etki')
plt.show()


```

# Data Preprocessing

```python
# Modelin egitimi uzerine etkisi olmayacak olan( olmayacagini yukaridaki iliskilere bakarak yorumladigim) sutunlarin modelin basarimini etkilemesin diye dusurulmesi

df.drop(columns=['Dogum_Tarihi','Uyruk','Il','Kullanici_id','Ilac_Baslangic_Tarihi','Ilac_Bitis_Tarihi','Yan_Etki_Bildirim_Tarihi'],inplace=True)
```

```python
df.head()
```

# Filling Missing Values

```python
df.isnull().sum() # dataframe'deki bos deger iceren sutunlarin bulunmasi ve ne kadar bos deger icerdiklerinin bilgisi
```

```python
#Bos degerlerin KNNImputer yardimi ile doldurulmasi

categorical_columns = ['Cinsiyet', 'Alerjilerim', 'Kronik Hastaliklarim', 
                       'Baba Kronik Hastaliklari', 'Anne Kronik Hastaliklari', 
                       'Kiz Kardes Kronik Hastaliklari', 'Erkek Kardes Kronik Hastaliklari', 'Kan Grubu','Ilac_Adi','Yan_Etki']
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str)) 
    label_encoders[col] = le  

imputer = KNNImputer(n_neighbors=3)
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```

```python
df_imputed.head()
```

# Encoding

```python
#Modelin veriyi anlayabilmesi icin kategorik veriler uzerine one-hot-encoding uygulanmasi

df_one_hot_encoded = pd.get_dummies(df_imputed, columns=categorical_columns, drop_first=True)
df_one_hot_encoded=df_one_hot_encoded.astype(int)
```

```python
df_one_hot_encoded.head()
```

```python
df_one_hot_encoded.isnull().sum()
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#Modelin egitiminin daha kisa surmesi ve bellek kullaniminin azaltilmasi icin normalizasyon isleminin uygulanmasi
columns_to_normalize=['Age','Kilo','Boy']
scaler=MinMaxScaler()
df_one_hot_encoded[columns_to_normalize]=scaler.fit_transform(df_one_hot_encoded[columns_to_normalize])
```
